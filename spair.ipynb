```python
import spair
import matplotlib.pyplot as plt
import cv2
```

```python
for annotation in spair.iterate_over_pairs('SPair-71k'):
    source = cv2.cvtColor(cv2.imread(annotation['source_path']), cv2.COLOR_BGR2RGB)
    target = cv2.cvtColor(cv2.imread(annotation['target_path']), cv2.COLOR_BGR2RGB)
    source_points = annotation['source_points']
    target_points = annotation['target_points']
    assert not annotation['mirror']
    print('Category:', annotation['category'])
    print('source_pose:', annotation['source_pose'], 'target_pose:', annotation['target_pose'])
    if set([annotation['source_pose'], annotation['target_pose']]) == {'Left', 'Right'}:
        # These are probably cases we will try to avoid.
        spair.plot_pair_correspondence(source, target, source_points, target_points, source_bounding_box=annotation['source_bounding_box'], target_bounding_box=annotation['target_bounding_box'])
        break
```
